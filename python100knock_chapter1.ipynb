データ観察

```python
# pandasライブラリのインポート
import pandas as pd
```

```python
# customer_masterの中身を確認
customer_master = pd.read_csv('customer_master.csv')
customer_master.head()
```

```python
# item_masterの中身を確認
item_master = pd.read_csv('item_master.csv')
item_master.head()
```

```python
# transaction_1の中身を確認
# 購買トランザクション
transaction_1 = pd.read_csv('transaction_1.csv')
transaction_1.head()
```

```python
# transaction_detail_1の中身を確認
# transaction_1の詳細データ
transaction_detail_1 = pd.read_csv('transaction_detail_1.csv')
transaction_detail_1.head()
```

```python
# transaction_2の中身を確認
# 購買トランザクション(transaction_1の続きみたいなイメージ)
transaction_2 = pd.read_csv('transaction_2.csv')
transaction_2.head()
```

```python
# transaction_detail_2の中身を確認
# transaction_2の詳細データ(transaction_detail_1の続きみたいなイメージ)
transaction_detail_2 = pd.read_csv('transaction_detail_2.csv')
transaction_detail_2.head()
```

データの結合

```python
# transaction_1とtransaction_2を縦積みで結合(union allみたいなイメージ)
transaction = pd.concat([transaction_1, transaction_2], ignore_index=True)
transaction.head()
```

```python
# 結合の確認
# transaction_1とtransaction_2のレコード数の和がtransactionのレコード数と等しいので、うまく結合できている
print(len(transaction_1))
print(len(transaction_2))
print(len(transaction))
```

```python
# transaction_detail_1とtransaction_detail_2を縦積みで結合
transaction_detail_2 = pd.read_csv('transaction_detail_2.csv')
transaction_detail=pd.concat([transaction_detail_1,transaction_detail_2], ignore_index=True)
transaction_detail.head()
```

```python
# 結合の確認
# transaction_detail_1とtransaction_detail_2のレコード数の和がtransactionのレコード数と等しいので、うまく結合できている
print(len(transaction_detail_1))
print(len(transaction_detail_2))
print(len(transaction_detail))
```

購買トランザクションの結合

```python
# pd.mergeにてテーブルの結合
# onにて結合キーを指定
# howにて結合方法の指定
join_data = pd.merge(transaction_detail, transaction[["transaction_id", "payment_date", "customer_id"]], on="transaction_id", how="left")
join_data.head()
```

```python
# join_dataのレコード数の確認
# transaction_detailのtransaction_idに対応していない部分があるため、join_dataにnull値あり
# なぜtransaction_idに抜けがあるのか確認★
print(len(transaction_detail))
print(len(transaction))
print(len(join_data))
```

購買データとマスタデータとの結合

```python
join_data = pd.merge(join_data, customer_master, on="customer_id", how="left")
join_data = pd.merge(join_data, item_master, on="item_id", how="left")
join_data.head()
```

priceカラムの作成

```python
# quantity*item_priceにてトランザクションにおけるpriceを算出
# 下記レコード4をみると、正確に計算されていることが分かる
join_data["price"] = join_data["quantity"] * join_data["item_price"]
join_data[["quantity", "item_price","price"]].head()
```

```python
# 作成したjoin_data["price"]とtransaction["price"]の合計値が等しいことを確認
# 期間をみて、どの程度の期間で売上が発生しているのかを確認してみる★
print(join_data["price"].sum())
print(transaction["price"].sum())
```

```python
# 以下のコードでも確認
join_data["price"].sum() == transaction["price"].sum()
```

```python
join_data.head()
```

```python
# 最新日の取得
latest_date = join_data["payment_date"].max()
# 最古日の取得
oldest_date = join_data["payment_date"].min()

print("最古日:", oldest_date)
print("最新日:", latest_date)

# つまり、約7ヶ月で971135000円の売上
```

各種統計量の確認

```python
# 各カラムのnull値の有無の確認
# null値は存在しないので、特に処理は必要ない
join_data.isnull().sum()
```

```python
# 各種統計量を確認
join_data.describe()

# 一度の購買におけるquantityは1つが多い
# ECサイトのデータだが、商品単価平均は121698
# 利用しているユーザは20〜80歳
```

月別でデータを集計してみる

```python
# 各カラムのデータ型の確認
# データ整形の際にエラーが出ることがあるので、必ず確認すること
join_data.dtypes
```

```python
# "payment_date"をdatetime型に変換
join_data["payment_date"] = pd.to_datetime(join_data["payment_date"])
# "payment_date"を文字型に変換しYYYYMMにする
join_data["payment_month"] = join_data["payment_date"].dt.strftime("%Y%m")
# 各月に変換ができていることを確認
join_data[["payment_date", "payment_month"]].head()
```

```python
# 各月ごとのpriceの合計値を算出
join_data.groupby("payment_month").sum()["price"]
```

月別、商品別でデータを集計してみる

```python
join_data.groupby(["payment_month","item_name"]).sum()[["price", "quantity"]]
```

```python
# pivot_tableにて行列の入れ替え
# BIツールに読み込む時などに便利そう
pd.pivot_table(join_data, index='item_name', columns='payment_month', values=['price', 'quantity'], aggfunc='sum')
```

売上推移を可視化してみる

```python
# indexにて軸の設定
graph_data = pd.pivot_table(join_data, index='payment_month', columns='item_name', values='price', aggfunc='sum')
graph_data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# plt.plotにて折れ線グラフの描画(Tableauであれば簡単に可視化できる)
plt.plot(list(graph_data.index), graph_data["PC-A"], label='PC-A')
plt.plot(list(graph_data.index), graph_data["PC-B"], label='PC-B')
plt.plot(list(graph_data.index), graph_data["PC-C"], label='PC-C')
plt.plot(list(graph_data.index), graph_data["PC-D"], label='PC-D')
plt.plot(list(graph_data.index), graph_data["PC-E"], label='PC-E')
plt.legend() 

# PC-Aの売上が常時高いことが分かる
```

```python
# indexにて軸の設定
graph_data_quantity = pd.pivot_table(join_data, index='payment_month', columns='item_name', values='quantity', aggfunc='sum')

plt.plot(list(graph_data_quantity.index), graph_data_quantity["PC-A"], label='PC-A')
plt.plot(list(graph_data_quantity.index), graph_data_quantity["PC-B"], label='PC-B')
plt.plot(list(graph_data_quantity.index), graph_data_quantity["PC-C"], label='PC-C')
plt.plot(list(graph_data_quantity.index), graph_data_quantity["PC-D"], label='PC-D')
plt.plot(list(graph_data_quantity.index), graph_data_quantity["PC-E"], label='PC-E')
plt.legend() 

# PC-Aは単価最も安いが、販売個数が最も多い
```
