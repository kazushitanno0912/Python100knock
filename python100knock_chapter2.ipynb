```python
# purchase_dateをdatetime型に変換
uriage_data["purchase_date"] = pd.to_datetime(uriage_data["purchase_date"])
# purchase_dateを各月に変換
uriage_data["purchase_month"] = uriage_data["purchase_date"].dt.strftime("%Y%m")
# aggfunc="size"で個数を集計
res = uriage_data.pivot_table(index="purchase_month", columns="item_name", aggfunc="size", fill_value=0)
res

# 99columnsより、表記揺れにより商品数が増幅している可能性がある
# ※商品Wと商品wがあるので
```

# 表記揺れを補正する

```python
# pd.uniqueにてitem_nameのユニーク数をカウント
print(len(pd.unique(uriage_data["item_name"])))


# この数値が実際の商品数と同等となるように修正する
```

```python
# upperで大文字指定
uriage_data["item_name"] = uriage_data["item_name"].str.upper()
# replaceでスペースを補完
uriage_data["item_name"] = uriage_data["item_name"].str.replace("　", "")
uriage_data["item_name"] = uriage_data["item_name"].str.replace(" ", "")
# item_nameでソートし出力
uriage_data.sort_values(by=["item_name"], ascending=True)
```

```python
# 改修後の表記の確認
print(pd.unique(uriage_data["item_name"]))
print(len(pd.unique(uriage_data["item_name"])))

# 実際の商品数が26であることを確認すること★
```

# 金額の欠損値を補完する

```python
# 欠損値の有無の確認
uriage_data.isnull().any(axis=0)
```

```python
# item_priceがnullのレコードを取り出す
flg_is_null = uriage_data["item_price"].isnull()
flg_is_null
```

```python
# for文で繰り返しtrgに値を入れつつ、その都度for文の中身を実行している、１つの値ごとに１回の処理が行われるイメージ
# locは条件を付与し、それに合致するデータを抽出することができる
for trg in list(uriage_data.loc[flg_is_null, "item_name"].unique()): #uriage_dataのitem_nameのうち、flg_is_nullと合致するitem_nameをユニークに取り出す
    price = uriage_data.loc[(~flg_is_null) & (uriage_data["item_name"] == trg), "item_price"].max() # 同じ商品で金額が正しく記載されている行を探し、その金額を取得している
    uriage_data.loc[(flg_is_null) & (uriage_data["item_name"]==trg),"item_price"] = price #欠損を起こしている商品名に対して、priceを代入している
uriage_data.head()
```

```python
# item_priceに欠損値がなくなったことを確認する
uriage_data.isnull().any(axis=0)
```

```python
# 欠損値の補完がうまくいっていることを見るために、商品ごとの最大、最小値をみる
for trg in list(uriage_data["item_name"].sort_values().unique()):
    print(trg + "の最大額：" + str(uriage_data.loc[uriage_data["item_name"]==trg]["item_price"].max()) + "の最小額：" + str(uriage_data.loc[uriage_data["item_name"]==trg]["item_price"].min(skipna=False)))
```

# 日付の表記揺れを修正

```python
# .str.isdigit()にてstr型を判定
flg_is_serial = kokyaku_data["登録日"].astype("str").str.isdigit()
flg_is_serial.sum()


#str型となっている日付が22箇所あることを確認
```

```python
kokyaku_data.head()
```

```python
# to_timedeltaでシリアル値をdatetimeに変換
fromSerial = pd.to_timedelta(kokyaku_data.loc[flg_is_serial, "登録日"].astype("float") - 2, unit="D") + pd.to_datetime('1900/1/1')
fromSerial
```

```python
# flg_is_serial以外のレコードをdatetimeに変換
fromString = pd.to_datetime(kokyaku_data.loc[~flg_is_serial, "登録日"])
fromString
```

```python
# concatでfromSerialとfromStringを結合し、kokyaku_data["登録日"]に代入
kokyaku_data["登録日"] = pd.concat([fromSerial, fromString])
kokyaku_data
```

```python
# kokyaku_data["登録日"]をYYYYMMの形にし登録年月として代入
kokyaku_data["登録年月"] = kokyaku_data["登録日"].dt.strftime("%Y%m")
rslt = kokyaku_data.groupby("登録年月").count()["顧客名"]
print(rslt)
print(len(kokyaku_data))
```
